```python
import pandas as pd

#Reading a csv file from github

url = "https://raw.githubusercontent.com/Impact-Insights/Hamoye-Internship/main/FoodBalanceSheets_E_Africa_NOFLAG.csv"
data = pd.read_csv(url, encoding='latin-1')
data
```

```python
# 1. Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on ‘Area’.
#   Which of these Areas had the highest sum in 2017?

data1 = data[["Area", "Y2017"]]
data1_1 = data1[data1["Area"] == "Nigeria"].groupby("Area")["Y2017"].sum()
print(data1_1)

data1_2 = data1[data1["Area"] == "Ethiopia"].groupby("Area")["Y2017"].sum()
print(data1_2)

data1_3 = data1[data1["Area"] == "South Africa"].groupby("Area")["Y2017"].sum()
print(data1_3)

data1_4 = data1[data1["Area"] == "Egypt"].groupby("Area")["Y2017"].sum()
print(data1_4)

data1_5 = data1[data1["Area"] == "Kenya"].groupby("Area")["Y2017"].sum()
print(data1_5, "\n", "\n")

all = pd.concat([data1_1,data1_2,data1_3,data1_4,data1_5])
print(all)

```

```python
#15 Perform a groupby operation on ‘Element’.
#   What year has the highest sum of Stock Variation?

years = ["Y2014", "Y2015", "Y2016", "Y2017", "Y2018"]
data15 = data[data["Element"] == "Stock Variation"].groupby("Element")[years].sum()
print(data15, "\n"*2)

import matplotlib.pyplot as plt

values = [58749.83, 34910.99, 33140.12, 54316.91, 20577.91]

plt.bar(years, values)
plt.xlabel('Year')
plt.ylabel('Stock Variation')
plt.title('Stock Variation Over Time')
plt.show()
```

```python
#16 What is the total number and percentage of missing data in 2014 to 3 decimal places?

data16 = data["Y2014"].isnull().sum()
print("Number of missing values in Y2014 is", data16, "\n")

perc = ((data16/len(data["Y2014"]))*100).round(3)
print("The percentage of missing values in Y2014 is", perc)
```
